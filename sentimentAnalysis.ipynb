```python
import numpy as np
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
# Extracting text from tweetList.csv
df = pd.read_csv('tweetlist.csv')
df = df.dropna(subset=["text"])
df1 = df.iloc[:, 1]
tweetList = [idx for idx in df1]
tweetList
```

```python
# Sentiment Analysis
pol_score = []
sub_score = []
for word in tweetList:
    print(word)
    blob = TextBlob(word)
    pol_score.append(blob.polarity)
    sub_score.append(blob.subjectivity)
    print(blob.sentiment)
```

```python
# Classification of polartiy
highNegative = []
highPozitive = []
negative = []
pozitive = []
notr = []

for score in pol_score:
    if score <= -0.5:
        highNegative.append(score)
    elif score < -0.1:
        negative.append(score)
    elif score <= 0.1:
        notr.append(score)
    elif score < 0.5:
        pozitive.append(score)
    else: 
        highPozitive.append(score

print(len(highNegative))
print(len(negative))
print(len(notr))
print(len(pozitive))
print(len(highPozitive))
print(len(pol_score))
```

```python
# Drawing Chart Belonging the Percentages of Clusters
import matplotlib.pyplot as plt

# Calculate percentages
numbers = [len(highNegative), len(negative), len(notr), len(pozitive), len(highPozitive)]
total = sum(numbers)
percentages = [(number/total)*100 for number in numbers]

labels = ['HighNeg', 'Neg', 'Notr', 'Pos', 'HighPos']
colors = ['red', 'purple', 'orange', 'lightgreen', 'cyan']
plt.pie(percentages, labels=labels, autopct='%1.1f%%', colors=colors)
plt.title('Percentage Distribution of Polarity Scores')
plt.axis('equal')
plt.show()
```

```python
# Plotting the dustrubion of subjectivity and polarity
plt.scatter(pol_score, sub_score)
plt.xlabel('Polarity')
plt.ylabel('Subjectivity')
plt.show
```

```python
avaragePol = sum(pol_score) / len(pol_score)
avaregeSub = sum(sub_score) / len(sub_score)
print('Avarege Polarity', avaragePol)
print('Avarege Subjectivity', avaregeSub)
```
